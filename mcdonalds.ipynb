```python
#Select columns 1 to 11 (0-based index) and convert to a NumPy array
md_x = data.iloc[:, 0:11]
```

```python
md_x
```

```python
# Convert "Yes" to 1 and everything else to 0
md_x = (md_x == "Yes").astype(int)
```

```python
md_x
```

```python
#Round the column means to two decimal places
rounded = md_x.mean().round(2)
```

```python
rounded
```

```python
from sklearn.decomposition import PCA
# Create a PCA model
pca = PCA()

# Fit the PCA model to your data
pca.fit(md_x)

# the standard deviations of principal components
std_dev = pca.explained_variance_ ** 0.5

# the proportion of variance explained by each component
variance_ratio = pca.explained_variance_ratio_

# Get the cumulative proportion of variance explained
cumulative_variance_ratio = variance_ratio.cumsum()

#  summary DataFrame
summary_df = pd.DataFrame({
    'Principal Component': range(1, len(std_dev) + 1),
    'Standard Deviation': std_dev,
    'Proportion of Variance Explained': variance_ratio,
    'Cumulative Proportion Explained': cumulative_variance_ratio
})

#  summary DataFrame
print(summary_df)
```

```python
MD_x_pca = pca.transform(md_x)

# Create a scatter plot of the PCA results
plt.scatter(MD_x_pca[:, 0], MD_x_pca[:, 1], c='grey', label='PCA Projection')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA Projection Scatter Plot')

# Show the plot
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

# Assuming 'MD_x' is your NumPy array with binary (0 and 1) data

# Initialize an empty list to store inertia values
inertia_values = []

# Range of clusters to consider (2 to 8)
num_clusters = range(2, 9)

# Loop through different numbers of clusters and calculate inertia
for k in num_clusters:
    kmeans = KMeans(n_clusters=k, random_state=1234)
    kmeans.fit(md_x)
    inertia_values.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(num_clusters, inertia_values, marker='o', linestyle='-', color='b')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia (Sum of Squared Distances)')
plt.title('Elbow Curve for K-Means Clustering')
plt.grid(True)
plt.show()

# Based on the plot, choose the optimal number of clusters where the curve starts to level off
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Create a pandas DataFrame from your data
data = {
    'yummy': [1, 0, 1, 0, 1, 0],
    'convenient': [1, 0, 0, 1, 1, 0],
    'spicy': [0, 1, 0, 1, 0, 1],
    'fattening': [0, 1, 0, 1, 0, 1],
    'greasy': [1, 0, 0, 1, 0, 1],
    'fast': [1, 0, 1, 0, 1, 0],
    'cheap': [1, 0, 1, 0, 0, 1],
    'tasty': [1, 0, 1, 0, 1, 0],
    'expensive': [0, 1, 0, 1, 0, 1],
    'healthy': [0, 1, 0, 1, 1, 0],
    'disgusting': [0, 1, 0, 1, 0, 1],
    'Like.n': [-5, -4, -3, -2, -1, 0]
}

df = pd.DataFrame(data)

# Define independent variables (X) and the dependent variable (y)
X = df[['yummy', 'convenient', 'spicy', 'fattening', 'greasy', 'fast', 'cheap', 'tasty', 'expensive', 'healthy', 'disgusting']]
y = df['Like.n']

# Create a Linear Regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Print the coefficients (regression coefficients)
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```
